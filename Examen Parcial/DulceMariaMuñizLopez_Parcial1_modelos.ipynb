```python
#Importar librerias 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
# Cargar los datos
df = pd.read_csv("Ecommerce Customers.csv")
# Gráficos de dispersión
plt.figure(figsize=(12, 5))
sns.pairplot(df, vars=["Length of Membership", "Yearly Amount Spent"])
plt.show()
```

```python
# Regresión Lineal entre Length of Membership y Yearly Amount Spent
X = df[["Length of Membership"]]
y = df["Yearly Amount Spent"]
```

```python
#Crear el modelo
model = LinearRegression()
model.fit(X, y)
#prediccion de cuanto gastara un cliente analmente cuando tenga 6.5 años como cliente 
years = np.array([[6.5]])
predicted_spent = model.predict(years)
print(f"Predicción de gasto anual para 6.5 años de membresía: {predicted_spent[0]:.2f}") 
```

```python
# Coeficiente de correlación y determinación
correlation = np.corrcoef(df["Length of Membership"], df["Yearly Amount Spent"])[0, 1]
r2 = r2_score(y, model.predict(X))
print(f"Coeficiente de correlación: {correlation:.4f}")
print(f"Coeficiente de determinación (R²): {r2:.4f}")
```

```python
# Gráfico de regresión
plt.figure(figsize=(8, 6))
sns.regplot(x=df["Length of Membership"], y=df["Yearly Amount Spent"], line_kws={"color": "red"})
plt.title("Regresión Lineal: Length of Membership vs Yearly Amount Spent")
plt.show()
```
