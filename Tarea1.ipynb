```python
import pandas as pd
import matplotlib.pyplot as plt

nombres = ["DULCE", "RODRIGO", "ALBERTO", "CAMILA"]
calificaciones = [[9, 8, 10], [10, 8.0, 7.5], [3.0, 8, 1], [10, 4.0, 5.0]]

df = pd.DataFrame({'Nombre': nombres, 'Calificaciones': calificaciones})

df['media'] = df['Calificaciones'].apply(lambda x: sum(x) / len(x))

print(df)

plt.bar(df['Nombre'], df['media'], color='blue')
plt.xlabel("Estudiantes")
plt.ylabel("Medias")
plt.title("Gr√°fico de barras con las medias de las calificaciones")

for idx, val in enumerate(df['media']):
    plt.text(idx, val + 0.1, f'{val:.2f}', ha='center', color="red", size=12)
plt.show()
```
