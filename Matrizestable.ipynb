```python
import numpy as np
import matplotlib.pyplot as plt
m1=np.matrix([[0.97, 0.02, 0.01], [0.6, 0.3, 0.1], [0.2, 0.1, 0.7]])
m1
```

```python
m_aux=np.matrix([[],[],[]])
m_aux_anterior=np.matrix([[],[],[]])
```

```python
for i in range(1000):
    if i==0:
        m_aux=np.dot(m1,m1)
        print(m_aux)
        print("----------", i+1)
        
    else:
        y=i+1
        m_aux_anterior=m_aux
        m_aux=np.dot(m_aux,m1) 
        print(m_aux)
        if (np.all(m_aux==m_aux_anterior) and (np.all(m_aux_anterior[:,0]==m_aux[:,0]))):
            break
        print(y, "----------", i+1)
print()
print(y,"la matriz estable es:\n", np.round(m_aux, 4))
m_aux=np.round(m_aux, 3)
```

```python
be=m_aux[0,0]
rmen=m_aux[0,1]
rmay=m_aux[0,2]
print(be, rmen, rmay)
```

```python
autos_taller=100*(rmen+rmay)
print("autos en el taller:", round(autos_taller,2))

#utilidad diaria
util_dia=(50*be)-(100*rmen)-(150*rmay)
util_anual=util_dia*365
print("utilidad diaria:", round(util_dia,4))
print("utilidad anual:", round(util_anual,4))
```
