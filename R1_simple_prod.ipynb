```python
#importar librerias 
import seaborn  as sns
import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_excel("produccion.xlsx",usecols="B,D")
df
```

```python
#selecionar las ejes o vars
variable_x="Horas TrabajadasX"
variable_y="Productos TerminadosY"
print("Variable X: ",variable_x)
print("Variable Y: ",variable_y)
 

```

```python
#generar analisis
modelo=LinearRegression()
modelo.fit(df[[variable_x]],df[variable_y])
print("Ecuacion de la recta es:y=",round(modelo.coef_[0],4),"x +",round(modelo.intercept_,4))
```

```python
#coeficiente de correlacion
print("Coeficiente de correlacion: ",round(np.corrcoef(df[variable_x],df[variable_y])[0,1],4))
```

```python
#coeficiente de determinacion
print("Coeficiente de determinacion: ",round(r2_score(df[variable_y],modelo.predict(df[[variable_x]])),4))
```

```python
#grafico con intervalo de confianza del 95%
sns.regplot(x=df[variable_x],y=df[variable_y],ci=95,line_kws={'color':'red'},scatter_kws={'color':'blue'})
plt.title("Regresion Lineal Simple")
plt.show()
```

```python
#prediccion horas
datos_predictor=55
horas_trabajadas_nuevas=pd.DataFrame([datos_predictor],columns=[variable_x])
autos_prod_prediccion=modelo.predict(horas_trabajadas_nuevas)
print("La prediccion de productos terminados para ",horas_trabajadas_nuevas.iloc[0][0],"horas trabajadas es: ",round(autos_prod_prediccion[0],4))
```
