```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Datos ficticios de clientes
data = {
    "Avg_Session_Length": np.random.uniform(5, 20, 50),  # Promedio de sesin en tienda
    "Time_on_App": np.random.uniform(10, 50, 50),  # Tiempo en la app
    "Time_on_Website": np.random.uniform(5, 30, 50),  # Tiempo en la web
    "Length_of_Membership": np.random.uniform(1, 10, 50),  # Aos como cliente
    "Yearly_Amount_Spent": np.random.uniform(200, 1000, 50)  # Gasto anual
}

df = pd.DataFrame(data)

# Grfico de dispersin
sns.pairplot(df)
plt.show()

# Seleccionar la variable predictora y la variable objetivo
X = df[["Length_of_Membership"]]  # Aos como cliente
y = df["Yearly_Amount_Spent"]  # Gasto anual

# Crear y entrenar el modelo de regresin lineal
model = LinearRegression()
model.fit(X, y)

# Predicciones y mtricas
y_pred = model.predict(X)
r2 = r2_score(y, y_pred)

# Coeficiente de correlacin y determinacin
correlation = np.corrcoef(df["Length_of_Membership"], df["Yearly_Amount_Spent"])[0, 1]
print(f"Coeficiente de correlacin: {correlation:.4f}")
print(f"Coeficiente de determinacin (R): {r2:.4f}")

# Prediccin para 6.5 aos como cliente
y_future = model.predict([[6.5]])
print(f"Gasto anual estimado para un cliente con 6.5 aos: ${y_future[0]:.2f}")

# Grfico de regresin
plt.figure(figsize=(8,5))
sns.regplot(x=X, y=y, line_kws={"color": "red"})
plt.xlabel("Aos como cliente")
plt.ylabel("Gasto anual")
plt.title("Regresin lineal: Aos de membresa vs Gasto anual")
plt.show()
```
