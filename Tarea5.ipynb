```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import linregress
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import statsmodels.api as sm
from sklearn.metrics import r2_score


# Mostrar las primeras filas de los datos y cargar datos
data = pd.read_csv("carros.csv")
print(data.head())
```

```python


# Crear gráfico de dispersión con regresión lineal
sns.lmplot(x="kms", y="precio", data=data, aspect=2, height=3)
plt.title("Gráfico de dispersión con regresión lineal")
plt.show()
```

```python
correlation = data["kms"].corr(data["precio"])
print(f"Coeficiente de correlación: {correlation}")
```

```python
r_squared = correlation ** 2
print(f"Coeficiente de determinación (R²): {r_squared}")
```

```python


#regresión lineal
slope, intercept, r_value, p_value, std_err = linregress(data["kms"], data["precio"])

#ecuación de la recta
print(f"Ecuación de la recta: Y = {slope}X + {intercept}")
```

```python
#predecir Y para X = 10
X_new = 10
Y_pred = slope * X_new + intercept
print(f"Predicción para X = {X_new}: Y = {Y_pred}")
```
