```python
#importar librerias
import pandas as pd
from math import sqrt
import numpy as np
import matplotlib.pyplot as plt
import os
import warnings
warnings.filterwarnings('ignore')
 
```

```python
#lectura de datos
#rutas
movies_df_filepath = os.path.abspath("datos/movies_original.csv")
ratings_df_filepath = os.path.abspath("datos/ratings_original.csv")
poster_df_filepath = os.path.abspath("datos/MovieGenre.csv")
 
#carga de datos
movies_df = pd.read_csv(movies_df_filepath)
ratings_df=pd.read_csv(ratings_df_filepath)
poster_df=pd.read_csv(poster_df_filepath, encoding='ISO-8859-1')
```

```python
movies_df.head()
```

```python
# Ajuste de datos 
movies_df['year'] = movies_df['title'].str.extract(r'(\(\d\d\d\d\))', expand=False)
print("extraer el año con parentesis", movies_df['year'])

# Quitar parentesis del año
movies_df['year'] = movies_df['year'].str.extract(r'(\d\d\d\d)', expand=False)
print("quitar parentesis del año", movies_df['year'])

movies_df['title'] = movies_df['title'].str.replace(r"\s+\(.*\)", "", regex=True)
```

```python
movies_df
```

```python
#Separar los valores de genero y generar un array 
movies_df['genres'] = movies_df['genres'].str.split('|')
movies_df.head()
```

```python
#Copiar el marco de datos de una copia 
movies_co= movies_df.copy()
```

```python
#colocar los valores de genero en la lista 1 y 0
for index, row in movies_df.iterrows():
    for genre in row['genres']:
        movies_co.at[index, genre] = 1
```

```python
#rellenar los valores NaN con 0
movies_co=movies_co.fillna(0)
movies_co
 
```

```python
#mostramos el rating de las peliculas
print("rating",ratings_df.head())
#eliminar la columna timestamp
ratings_df=ratings_df.drop('timestamp',axis=1)
print("rating sin timestamp:\n",ratings_df.head())
```

```python
#definamos de forma constante las preferencias del usuario
usuario_en=[
    {'title':'Breakfast Club, The','rating':5},
    {'title':'Toy Story','rating':3.5},
    {'title':'Jumanji','rating':2},
    {'title':"Pulp Fiction",'rating':5},
    {'title':'Akira','rating':4.5}
]
#lo convertimos en un dataframne
entrada_peli = pd.DataFrame(usuario_en)
entrada_peli
```

```python
#vamos a agregar el id a nuestras peliculas de entrada
Id = movies_df[movies_df['title'].isin(entrada_peli['title'].tolist())]
entrada_peli = pd.merge(Id, entrada_peli, how='inner', on='title', validate='many_to_many')
print("peliculas de entrada con el id:\n", entrada_peli.head())
 
```

```python
#Eliminamos informacion necesara
entrada_peli=entrada_peli.drop('genres',axis=1).drop('year',axis=1)
print("Ya con las columnas eliminadas",entrada_peli.head())
```

```python
#metodo  one hot encoding
peli_usuario=movies_co[movies_co['movieId'].isin(entrada_peli['movieId'].tolist())]
print("peliculas de usuario codificadas:\n", peli_usuario)
```

```python
#Seguimos limpiando
peli_usuario=peli_usuario.reset_index(drop=True)
tabla_generos=peli_usuario.drop('movieId',axis=1).drop('title',axis=1).drop('genres',axis=1).drop('year',axis=1)
print("tabla de generos: \n", tabla_generos)
```

```python
#Hacer la matris de paso y utilizar la funcion hot
print("Tabla de generos transpuestas: \n",tabla_generos.transpose())
print("\n")
print("Datos para ver rating :",entrada_peli["rating"])
```

```python
#uasamos dot
perfil_usu = tabla_generos.transpose().dot(entrada_peli["rating"])
print("Categoria que el usuario prefiere:", perfil_usu)
```

```python
#Mostramos tabla original de generos 
generos = movies_co.set_index('movieId')
generos
                            
                            
```

```python
#Mostrar informacion necesaria 
generos = generos.drop('title', axis=1).drop('genres', axis=1).drop('year', axis=1)
print("Generos", generos.head(20))
generos.shape
```

```python
#ahota realizamos el promedio ponderado para recomendar 20 pelis de la bd
#y creamo ponderation cmo estructura del df
generos_reset = generos.reset_index()
recom = pd.DataFrame(generos_reset['movieId'])
recom['ponderations'] = (generos_reset.multiply(perfil_usu, axis=1).sum(axis=1)) / (perfil_usu.sum())
recom = pd.DataFrame(recom)
recom.head(20)
```

```python
#Ordenamos tabla de ponderacion 
recom = recom.sort_values('ponderations', ascending=False)
recom = recom.head(20)
```

```python
#tabla final d recom
movies_df.set_index('movieId')
recom.set_index('movieId')
 
movies_df_position=movies_df.join(recom,lsuffix='_key',on='movieId',how='inner',validate='many_to_many')
movies_df_position=movies_df_position.sort_values('ponderations',ascending=False)
movies_df_position.head(20)
```
